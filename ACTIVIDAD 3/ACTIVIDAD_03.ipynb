**Modulo para calcular la frecuencia get_frecuent_itemset**




```python
def get_frecuent_itemset(playlist,min_support):
    #calculando lista cocn elementos de minimo soporte
    all_list=[]
    for i in range(len(playlist)):
        if len(playlist[i])>min_support:
            all_list.append(playlist[i])

    #calculando la frecuencia de caniones unicas
    diccionario_frecuencias = {}
    for i in range(len(playlist)):
        for j in range(len(playlist[i])):
            palabra=playlist[i][j]
            if palabra in diccionario_frecuencias:
                diccionario_frecuencias[palabra] += 1
            else:
                diccionario_frecuencias[palabra] = 1

    print("-- FRECUENCIA HALLADA EXITOSAMENTE --")
    return all_list,diccionario_frecuencias
```

Para el soporte tomaremos un umbral de 0.015 respecto a cada cancion esto para limitar y tener un menor data set, y ademas porque las canciones con un soporte bajo no seran relevantes para la creacion de reglar en la que nos basaremos

```python
def hallar_soporte(dic_frec):
    diccionario_soporte = {}
    for palabra in dic_frec:
        #print(dic_frec[palabra])
        if(dic_frec[palabra]/len(data)>=confianza_min):
            diccionario_soporte[palabra]=dic_frec[palabra]/len(data)
    return diccionario_soporte
```

# 4. Creacion de reglas
## - Soporte
## - Confianza
## - Lift

```python
def generate_association_rules(frequent_itemsets, confidence = 0.2, lift =1.5):
    tup=list(frequent_itemsets.keys())
    #print(len(tup))
    rules = [[0 for x in range(2)] for y in range(len(tup))] 
    for i in range(len(rules)):
        for j in range(2):
            #print(i,j)
            rules[i][j]=tup[i][j]
    # calculamos confianza
    dict_confianz=calcular_confianza(dict_soporte,dict_soport_tuplas)
    # calculamos lift
    dict_lift=calcular_lift(dict_soporte,dict_confianz)
    # validamos con confidence y lift
    tup=list(frequent_itemsets.keys())
    rules_final = []
    for k in range(len(tup)):
        #print(tup[k][0])
        #print(tup[k])
        conf_act=dict_confianz[tuple(tup[k])]
        lift_act=dict_lift[tuple(tup[k])]
        if (conf_act>=confidence) and  (lift_act>=lift):
            rules_final.append([tup[k],conf_act,lift_act])
    
    return rules,dict_confianz,dict_lift,rules_final
def calcular_confianza(dict_soporte,dict_soport_tuplas):
    dict_confianza={}
    tup=list(dict_soport_tuplas.keys())
    for k in range(len(tup)):
        #print(tup[k][0])
        #print(tup[k])
        dict_confianza[tup[k]]=dict_soport_tuplas[tuple(tup[k])]/(dict_soporte[tup[k][0]])
    return dict_confianza

def calcular_lift(dict_soporte,dict_confianz):
    dict_lift={}
    tup=list(dict_confianz.keys())
    for k in range(len(tup)):
        #print(tup[k][0])
        #print(tup[k])
        dict_lift[tup[k]]=dict_confianz[tuple(tup[k])]/(dict_soporte[tup[k][1]])
    return dict_lift
```
